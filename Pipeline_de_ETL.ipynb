```python
import pandas as pd
from google.colab import files
import io

# upload do arquivo CSV
uploaded = files.upload()

# Identificação do arquivo CSV
nome_arquivo = 'produtividade3.csv'

# Carregando os dados em um DataFrame
df = pd.read_csv(io.StringIO(uploaded[nome_arquivo].decode('utf-8')))

# Transformação desejada (multiplicar a coluna "qtd_peca_produzida" por 2)
df['qtd_peca_produzida'] = df['qtd_peca_produzida'] * 2

# Salvar os dados transformados em um novo arquivo CSV
nome_arquivo_saida = "produtividade_transformado.csv"
df.to_csv(nome_arquivo_saida, index=False)
```
